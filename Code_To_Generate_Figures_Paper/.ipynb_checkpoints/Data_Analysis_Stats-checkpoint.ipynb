```python
#### Chi Square Function
class ChiSquare:
    def __init__(self, dataframe):
        self.df = dataframe
        self.p = None #P-Value
        self.chi2 = None #Chi Test Statistic
        self.dof = None
        
        self.dfObserved = None
        self.dfExpected = None
        
    def _print_chisquare_result(self, colX, alpha):
        result = ""
        if self.p<alpha:
            result="{0} is statistically significant and P value is {1}".format(colX,self.p)
        else:
            result="{0} is NOT statistically significant  and P value is {1}. (Discard {0} from model)".format(colX,self.p)
        print(result)
        
    def TestIndependence(self,colX,colY, alpha=0.05):
        X = self.df[colX].astype(str)
        Y = self.df[colY].astype(str)
        
        self.dfObserved = pd.crosstab(X,Y) 
        print(self.dfObserved)
        chi2, p, dof, expected = stats.chi2_contingency(self.dfObserved.values)
        print("Expected table")
        print(expected)
        
        if np.min(expected)< 5:
            print('Condition for Chi square is fail')
        else:
            print('Condition for Chi square is accepted')
        self.p = p
        self.chi2 = chi2
        self.dof = dof 
        
        self.dfExpected = pd.DataFrame(expected, columns=self.dfObserved.columns, index = self.dfObserved.index)
        
        self._print_chisquare_result(colX,alpha)
        
 
```

```python
#### Fisher test function
class fisher_test:
    def __init__(self, dataframe):
        self.df = dataframe
        self.p = None #P-Value
        self.ODD_ratio = None #Odd ratio of fisher test      
        self.dfObserved = None
        
        
    def _print_fisher_result(self, colX, alpha):
        result = ""
        if self.p<alpha:
            result="{0} is statistically significant and P value is {1}".format(colX,self.p)
        else:
            result="{0} is NOT statistically significant  and P value is {1}".format(colX,self.p)
        print(result)
        
    def TestIndependence(self,colX,colY, alpha=0.05):
        X = self.df[colX].astype(str)
        Y = self.df[colY].astype(str)
        
        self.dfObserved = pd.crosstab(X,Y) 
        print(self.dfObserved)
        oddsratio, p = stats.fisher_exact(self.dfObserved.values)
        self.p = p
        self.ODD_ratio = oddsratio     
        self._print_fisher_result(colX,alpha)
 
```

```python
# Encode a numeric column as zscores
def encode_numeric_zscore(df, name, mean=None, sd=None):
    if mean is None:
        mean = df[name].mean()

    if sd is None:
        sd = df[name].std()

    df[name] = (df[name] - mean) / sd
    
# Encode text values to dummy variables(i.e. [1,0,0],[0,1,0],[0,0,1] for red,green,blue)
def encode_text_dummy(df, name):
    dummies = pd.get_dummies(df[name])
    for x in dummies.columns:
        dummy_name = f"{name}_{x}"
        df[dummy_name] = dummies[x]
    df.drop(name, axis=1, inplace=True)
```

```python
# Encode a numeric column as zscores
def encode_numeric_zscore_train(df, name, mean=None, sd=None):
    if mean is None:
        mean = df[name].mean()

    if sd is None:
        sd = df[name].std()

    df.loc[:,name] = (df.loc[:,name] - mean) / sd
    return mean,sd

def encode_numeric_zscore_test(df, name, mean_train, sd_train):

    df.loc[:,name] = (df.loc[:,name] -mean_train) / sd_train
    
       
# Encode text values to dummy variables(i.e. [1,0,0],[0,1,0],[0,0,1] for red,green,blue)
def encode_text_dummy(df, name):
    dummies = pd.get_dummies(df[name])
    for x in dummies.columns:
        dummy_name = f"{name}_{x}"
        df[dummy_name] = dummies[x]
    df.drop(name, axis=1, inplace=True)
```

### Function for calculating median and interquatile range for total sub group

```python
def stats_covariate_totalgroup(data_subgroup,covariate):
    data_subgroup[covariate].describe()
    X_med = np.percentile(data_subgroup[covariate], 50, axis =0)
    ### Calculating 95 CI
    alpha = 0.95
    #### Lower interval
    #p_L = ((1.0-alpha)/2.0) * 100 # 2.5
    X_L = np.percentile(data_subgroup[covariate], 25, axis =0)
    ### Upper interval 
    #p_U = (alpha+((1.0-alpha)/2.0)) * 100 # 97.5
    X_U = np.percentile(data_subgroup[covariate], 75 , axis =0)
    # covariate mean
    Cov_mean = np.mean(data_subgroup[covariate])
    Cov_std = np.std(data_subgroup[covariate],ddof=1)
    print('Median,25%,75% is',X_med,X_L,X_U )
    print('Mean, Standard deviation',Cov_mean,Cov_std)
```

### Function for calculating median and interquatile range for the sub group condition

```python
def stats_covariate_condition(data_subgroup,covariate,condition_covariate,condition_val):
    X_med = np.percentile(data_subgroup.loc[data_subgroup[condition_covariate]==condition_val,covariate], 50, axis =0)
    ### Calculating 95 CI
    alpha = 0.95
    #### Lower interval
    #p_L = ((1.0-alpha)/2.0) * 100 # 2.5
    X_L = np.percentile(data_subgroup.loc[data_subgroup[condition_covariate]==condition_val,covariate], 25, axis =0)
    ### Upper interval 
    #p_U = (alpha+((1.0-alpha)/2.0)) * 100 # 97.5
    X_U = np.percentile(data_subgroup.loc[data_subgroup[condition_covariate]==condition_val,covariate], 75 , axis =0)
    # Mean 
    cov_mean = np.mean(data_subgroup.loc[data_subgroup[condition_covariate]==condition_val,covariate])
    # Standard deviation
    cov_std = np.std(data_subgroup.loc[data_subgroup[condition_covariate]==condition_val,covariate],ddof=1)
    print('Median,25%,75% is',X_med,X_L,X_U )
    print('Mean, Standard deviation',cov_mean,cov_std)
```

```python
alpha = 0.05
stat, p = mannwhitneyu(data_subgroup.loc[data_subgroup['NSAED_SAED_Increase_Exposure_After_EEG']==0,'Post_Treat_Days'], data_subgroup.loc[data_subgroup['NSAED_SAED_Increase_Exposure_After_EEG']==1,'Post_Treat_Days'])
if p > alpha:
    result="Same distribution and P value is {0}".format(p)
    print(result)
    #print('Same distribution (fail to reject H0)')
else:
    result="Different distribution and P value is {0}".format(p)
    print(result)
```
